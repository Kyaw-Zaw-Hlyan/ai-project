```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/2nd year/tri 3/AI/project/polymers.csv')

# Print column names to verify
print(data.columns)

# Prepare the data
X = data.drop(columns=['Material'])
y = data['Material']

# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Encode the target labels
y_train_encoded = pd.get_dummies(y_train).values
y_test_encoded = pd.get_dummies(y_test).values

# Build the neural network
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(y_train_encoded.shape[1], activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train_encoded, epochs=100, batch_size=10, validation_split=0.2)


```

```python
def get_user_input():
    properties = [' modulus_of_elasticity', 'tensile_Strength', 'thermal_conductivity', 'density', 'cost/kg']
    user_input = {}

    for prop in properties:
        value = input(f"Enter desired {prop.replace('_', ' ')} (or skip): ")
        if value.strip().lower() != 'skip':
            try:
                user_input[prop] = float(value)
            except ValueError:
                print("Invalid input, please enter a valid number.")
                return get_user_input()

    return user_input

def find_nearest_polymers(model, scaler, data, user_input):
    # Prepare the user input
    input_data = []
    for column in data.drop(columns=['Material']).columns:
        if column in user_input:
            input_data.append(user_input[column])
        else:
            input_data.append(np.nan)

    # Handle missing values
    input_data = pd.DataFrame([input_data], columns=data.drop(columns=['Material']).columns)
    input_data = input_data.fillna(data.drop(columns=['Material']).mean())
    input_data_scaled = scaler.transform(input_data)

    # Predict
    predictions = model.predict(input_data_scaled)
    predicted_probabilities = np.max(predictions, axis=1)

    # Find the top N predictions
    top_indices = np.argsort(-predicted_probabilities)
    top_polymers = data.iloc[top_indices]

    return top_polymers

def main():
    user_input = get_user_input()
    top_polymers = find_nearest_polymers(model, scaler, data, user_input)

    print("Top polymer based on your preferences:")
    print(top_polymers[['Material', ' modulus_of_elasticity', 'tensile_strength', 'thermal_conductivity', 'density', 'cost/kg']])

if __name__ == "__main__":
    main()
```
